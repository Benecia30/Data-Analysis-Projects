# ML Algorithm 

## DECISION TREE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
from sklearn.preprocessing import StandardScaler
import warnings
```

```python
data_encoded.columns
```

```python

# Identify the target variable
target_variable = 'Particulars_Net_profit'
target = data_encoded[target_variable]

# Select features
features = data_encoded.drop(target_variable, axis=1)
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Preprocess numerical features - Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize the Decision Tree classifier
dt_classifier = DecisionTreeClassifier(max_depth=3)
```

```python
# Fit the model to the scaled training data (reshape X_train)
dt_classifier.fit(X_train_scaled, y_train)
```

```python
# Make predictions on the scaled test data
predictions = dt_classifier.predict(X_test_scaled)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')
```

```python

class_weights ='balanced'
dt_classifier_weighted = DecisionTreeClassifier(max_depth=3, class_weight=class_weights)
dt_classifier_weighted.fit(X_train_scaled, y_train)
predictions_weighted = dt_classifier_weighted.predict(X_test_scaled)
```

```python
report = classification_report(y_test, predictions,zero_division=1)
print("Classification Report:\n", report)
```

```python
plt.figure(figsize=(10, 10))
plot_tree(dt_classifier_weighted, filled=True, feature_names=features.columns.tolist(), rounded=True, fontsize=12)
plt.show()


```
