```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defining custom autopct function for formatting percentage values 
def custom_autopct(value):
    return f'{value:.2f}%'
```

```python
#Data for pie chart
labels=['UK','Canada','India','Russia','China','Australia','USA','France']
values=np.array([29.75,11,3,7,2.5,10.5,19.25,17])

#Figure size with specified data
plt.figure(figsize=(8, 8))

#Plotting the pie chart with default colors, labels and autopct
plt.pie(values,labels=labels,autopct=custom_autopct,startangle=90)

#Set equal aspect ratio to ensure the pie chart is circular
plt.axis('equal')

#Adjusting the layout to provide space for the title
plt.subplots_adjust(left=0.1,right=0.9,top=0.9,bottom=0.1)

#Adding a title to the pie chart
plt.title('Piechart of Oct 2023')

#Display the pie chart
plt.show()
```

## TURNOVER OF EACH COUNTRY

```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
values=np.array([29.75,11,3,7,2.5,10.5,19.25,17])

#Total income
total_income = 2118077689  

# Calculate individual income for each country
individual_income = (values / 100) * total_income

# Create a list of tuples containing country and income
country_income_list = list(zip(['UK', 'Canada', 'India', 'Russia', 'China', 'Australia', 'USA', 'France'], individual_income))

# Sort the list in descending order based on income
sorted_country_income = sorted(country_income_list, key=lambda x: x[1], reverse=True)

# Display the sorted individual income for each country
for country, income in sorted_country_income:
    print(f'{country}: ${income:.2f}')
    

# Define different colors for each country
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightskyblue', 'lightpink', 'lightgreen', 'lightsteelblue']

#Figure size with specified data
plt.figure(figsize=(10, 6))

# Create a bar graph
plt.bar(['UK', 'Canada', 'India', 'Russia', 'China', 'Australia', 'USA', 'France'], individual_income, color=colors)
plt.xlabel('Country')
plt.ylabel('Income')
plt.title('Income by Country')

# Display income values on top of each bar
for index, value in enumerate(individual_income):
    plt.text(index, value, f'${value:.2f}', ha='center', va='bottom')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')  

plt.show()
```

## EXPENSE OF THE UK

```python
#Given Values
total_expense_uk=2231721.54
total_expense_company=7501585
marketing_expense_company=1500000
collab_expense_company=1000000
salary_expense_company=4500000
other_expense_company=501585

#Calculate the divided expense for the UK
marketing_expense_uk=(marketing_expense_company*total_expense_uk)/total_expense_company
collab_expense_uk=(collab_expense_company*total_expense_uk)/total_expense_company
salary_expense_uk=(salary_expense_company*total_expense_uk)/total_expense_company
other_expense_uk=(other_expense_company*total_expense_uk)/total_expense_company

#displaying individual expense of marketing,collab,salary,other of the UK
print('Marketing expense of the UK:','$',marketing_expense_uk)
print('Collaboration expense of the UK:','$',collab_expense_uk)
print('Salary expense of the UK:','$',salary_expense_uk)
print('Other expense of the UK:','$',other_expense_uk)

# Plotting a bar graph
expenses = [marketing_expense_uk, collab_expense_uk, salary_expense_uk, other_expense_uk]
expense_labels = ['Marketing', 'Collaboration', 'Salary', 'Other']

# Display Expense values on top of each bar and plotting the bar graph
for index, value in enumerate(expenses):
    plt.text(index, value, f'${value:.2f}', ha='center', va='bottom')
    
plt.bar(expense_labels, expenses, color=['blue', 'green', 'orange', 'red'],width=0.3)
plt.xlabel('Expense Type')
plt.ylabel('Expense Amount ($)')
plt.title('Individual Expenses for the UK')
plt.show()
```

```python
values=np.array([32,2,2.3,6,1.70,8,20,28])

#Total expense
total_expense = 9001585  

# Calculate individual expense for each country
individual_expense = (values / 100) * total_expense

# Create a list of tuples containing country and expense
country_expense_list = list(zip(['UK', 'Canada', 'India', 'Russia', 'China', 'Australia', 'USA', 'France'], individual_expense))

# Sort the list in descending order based on expense
sorted_country_expense = sorted(country_expense_list, key=lambda x: x[1], reverse=True)

# Display the sorted individual expense for each country
for country, expense in sorted_country_expense:
    print(f'{country}: ${expense:.2f}')
    
# Define different colors for each country
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightskyblue', 'lightpink', 'lightgreen', 'lightsteelblue']

#Figure size with specified data
plt.figure(figsize=(10, 6))

# Create a bar graph
plt.bar(['UK', 'Canada', 'India', 'Russia', 'China', 'Australia', 'USA', 'France'], individual_expense, color=colors)
plt.xlabel('Country')
plt.ylabel(' Expense')
plt.title( 'Expense by Country')

# Display Expense values on top of each bar
for index, value in enumerate(individual_expense):
    plt.text(index, value, f'${value:.2f}', ha='center', va='bottom')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')  

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

```

```python
# Assuming you have a dataset with income and expense data for October and November
# Replace this with your actual dataset or use the provided profit data

# Given profit values for each country in October and November 2023
profits_october = {
    'UK': 627896390.94,
    'Canada': 232163371.44,
    'India': 63317283.12,
    'Russia': 147740327.28,
    'China': 52764402.61,
    'Australia': 221610490.92,
    'USA': 406285900.02,
    'France': 358797937.68
}

profits_november = {
    'UK': 607383745.28,
    'Canada': 37961484.08,
    'India': 43655706.70,
    'Russia': 113884452.24,
    'China': 32267261.46,
    'Australia': 151845936.32,
    'USA': 379614840.80,
    'France': 531460777.12
}

# Create a DataFrame from the profit data
df = pd.DataFrame({
    'Country': list(profits_october.keys()),
    'Income_October': list(profits_october.values()),
    'Income_November': list(profits_november.values())
})

```

```python
# Combine October and November data to predict December
df['Income_December_Prediction'] = (df['Income_October'] + df['Income_November']) / 2  # Simple average as a basic prediction

```

```python
# Split the data into features (X) and target variable (y)
X = df[['Income_October', 'Income_November']]
y = df[['Income_December_Prediction']]

```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```
